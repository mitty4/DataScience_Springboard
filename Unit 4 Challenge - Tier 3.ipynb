**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

Don't be afraid to use StackOverflow for help  with this.

```python
#transpose the dataframe, which swaps the columns with the rows. Don't forget to assign the transposed df.
p_t = properties.T
#reset the index to zero-indexed list of integers
p_t = p_t.reset_index()
#reassign the column names as the first row in the df
p_t.columns = p_t.iloc[0]
#drop the row that now matches the column names
p_t = p_t.drop(0)
#view changes
p_t.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
#the first two column names are not appropriately descriptive so rename them
p_t = p_t.rename(columns={'Unnamed: 0':'London_Borough', pd.NaT:'ID'}, errors="raise")
p_t.head(10)
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
#now lets tidy the data to be longer than is wide by melting the times(Month) into one column
clean_p = pd.melt(p_t, id_vars=['London_Borough','ID'])
#rename the columns appropriately
clean_p = clean_p.rename(columns={0:'Month','value':'Average Price'})
clean_p.head()
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
#Sweet! Take an inside look at the type of the data in each column
clean_p.dtypes
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
#are all the columns equal in length?
clean_p.count()
#how many distinct/unique instances of london boroughs are there?
clean_p['London_Borough'].unique()
#drop any(default) rows that have a na value and assign it to a variable
NaNFreeDF2 = clean_p.dropna()
#are the columns equal now?
NaNFreeDF2.count()
#define the boroughs that are not part of the study yet made their way into the data anyway (not sure how to find this other than it being given)
nonBoroughs = ['Inner London', 'Outer London', 
               'NORTH EAST', 'NORTH WEST', 'YORKS & THE HUMBER', 
               'EAST MIDLANDS', 'WEST MIDLANDS',
              'EAST OF ENGLAND', 'LONDON', 'SOUTH EAST', 
              'SOUTH WEST', 'England']
#define our clean and workable df by negating the nonBoroughs list within NanFreeDF2
df = NaNFreeDF2[~NaNFreeDF2.London_Borough.isin(nonBoroughs)]
#now how many entries are there?
df.count()
#and shape?
df.shape
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
#convert the object to a float for aggregation later
df['Average Price'] = pd.to_numeric(df['Average Price'])
#get the average, average price by borough for each year and reset the index
dfg = df.groupby(by=['London_Borough', 'Year'])[['Average Price']].mean().reset_index()
dfg.head()
```

```python
#create a function to find the ratio change of average house price
def avg_price_ratio(d):
    y1998 = float(d['Average Price'][d['Year']==1998])
    y2018 = float(d['Average Price'][d['Year']==2018])
    ratio = [y2018/y1998]
    return ratio

#initialize an empty dictionary to accept key, value average price by borough pairs 
final = {}
#loop through each distinct borough in the df
#get the individual row
#call avg_price_ratio on each borough row and add the result to final dict
for b in dfg['London_Borough'].unique():
    borough = dfg[dfg['London_Borough']==b]
    final[b]=avg_price_ratio(borough)

#create a dataframe with the populated dictionary
df_ratios = pd.DataFrame(final)

#Lots going on in this row. but nothing we haven't seen before.
#transpose the df, reset the index, rename the columns and assign it to itself
df_ratios = df_ratios.T.reset_index().rename(columns={'index':'London_Borough', 0:'Avg_price_ratio'})

#get a snippet of the df
#get the top 15 boroughs and sort them high to low
top15 = df_ratios.sort_values(by='Avg_price_ratio', ascending=False).head(15)
```

```python
#create bar plot with borough and price ratio columns
ax = top15[['London_Borough','Avg_price_ratio']].plot(kind='bar')

#set the x ticks and y ticks accordingly
ax.set_xticklabels(top15.London_Borough)
ax.set_yticklabels(top15.Avg_price_ratio)
```
